```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
website = 'https://www.cars.com/shopping/results/?stock_type=all&makes%5B%5D=bmw&models%5B%5D=&list_price_max=&maximum_distance=20&zip='
```

```python
response = requests.get(website)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content,'html.parser')
```

```python
soup
```

```python
result = soup.find_all('div',{'class':'vehicle-card'})
```

```python
name = []
mileage = []
sds_rating__count = []
sds_rating__link = []
primary_price = []
dealer_name = []
```

```python
for item in result:
    try:
        name.append(item.find('h2').get_text())
    except:
        name.append('n/a')
    try:
        mileage.append(item.find('div',{'class':'mileage'}).get_text())
    except:
        mileage.append('n/a')
    try:
        sds_rating__count.append(item.find('span',{'class':'sds-rating__count'}).get_text())
    except:
        sds_rating__count.append('n/a')
    try:
        sds_rating__link.append(item.find('span',{'class':'sds-rating__link'}).get_text())
    except:
        sds_rating__link.append('n/a')
    try:
        primary_price.append(item.find('span',{'class':'primary-price'}).get_text())
    except:
        primary_price.append('n/a')
    try:
        dealer_name.append(item.find('div',{'class':'dealer-name'}).get_text().strip())
    except:
        dealer_name.append('n/a')
```

```python
car = pd.DataFrame({'name':name,
                  'mileage':mileage,
                  'stars':sds_rating__count,
                  'reviews':sds_rating__link,
                  'price':primary_price,
                  'dealer_name':dealer_name})
```
